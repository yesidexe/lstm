Creo una columna con el nombre `datetime` a la que le doy formato de fecha, sacado de la columna `Date Time`, la pongo como index, ordeno y elimino.

```python
df['datetime'] = pd.to_datetime(df['Date Time'],format='%d.%m.%Y %H:%M:%S')
```

```python
# Hay que poner 'datetime' y no df['datetime'] para que agarre la columna y no cree una copia y deje la otra en el df.
df = df.set_index('datetime')
df = df.sort_index()
df = df.drop('Date Time', axis=1)
```

## EDA

```python
cols = df.columns
N = len(df)
plots = df[cols][0:N]
plots.index = df.index[0:N]
_ = plots.plot(subplots=True, figsize=(12, 16))
```
